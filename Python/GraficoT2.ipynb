```python
import matplotlib.pyplot as plt
import numpy as np

w1=[]
z1=[]
z2=[]



dados2 = open("/home/user/pasta/Documentos/TCC/Dif-2D/new/Fortran/converg3.txt" , "r")
for line in dados2:
    line = line.strip()
    Nr , EMax , Emed = line.split()
    nr=float(Nr)
    emax=float(EMax)
    emed=float(Emed)
    w1.append(nr)
    z1.append(emax)
    z2.append(emed)
dados2.close()


plt.plot(w1,z1,linestyle='-')
#plt.plot(w1,z2,linestyle='-')
plt.title(""r"$Converg\^encia$")
plt.ylabel(""r"$\frac{T_{N}-T_{a}}{T_{a}}$" ,  fontsize=18)
plt.yscale("log")
plt.xscale("log")
plt.xlabel(""r"$N_{r}$" ,  fontsize=12)
plt.tight_layout()
plt.legend([r"$Erro\;\;M\'aximo$",r"$Erro\;\;M\'edio$"],  fontsize=10)
plt.grid()
plt.savefig('Convergencia.png')
plt.show()
```

```python
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np
import csv
import scipy.interpolate
t, r, z = np.loadtxt("/home/user/pasta/Documentos/TCC/Code-TCC/Fortran/Temp_2d.txt",delimiter= ";",unpack=True)

ordem=int(np.size(r)**0.5)
T = t.reshape(ordem,ordem)
R = r.reshape(ordem,ordem)*100 
Z = z.reshape(ordem,ordem)*100 

fig,ax=plt.subplots(1,1)

cp = ax.contourf(Z,R,T)
fig.colorbar(cp)
ax.set_aspect("equal")
plt.title(r"$Distribuic\c\~ao\;\;de\;\;Temperatura - N=50$")
plt.ylabel("r[cm]")
plt.xlabel("z[cm]")
plt.tight_layout()
#plt.legend([r"$Num\'erico\;\;  N_{r}=90$"])
plt.savefig('T_2d-cT-N50.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=[]
y1=[]


dados = open("/home/user/pasta/Documentos/TCC/Code-TCC/Fortran/Temp_2d_su.txt" , "r")
for line in dados:
    line = line.strip()
    TN , Z = line.split()
    tn=float(TN)
    z=float(Z)*100
    x.append(z)
    y1.append(tn)
dados.close()


plt.plot(x,y1,linestyle='-',marker='x',markersize=2)

plt.title(r"$Distribuic\c\~ao\;\;de\;\;Temperatura $")
plt.ylabel("T[C]")
plt.xlabel("z[cm]")
plt.legend([r"$Num\'erico\;\;  N_{r}=20$",r"$Num\'erico\;\;  N_{r}=10$",r"$Anal\'itico$",])
plt.grid()


plt.savefig('T_surf.png')
plt.show()
```
